###Building Percussive Dataset

```python


# set input and output directories
input_dir = '/content/drive/MyDrive/GTZAN/Demux'
output_dir = '/content/drive/MyDrive/GTZAN/PERCU'

# iterate through the files in the input directory
for genre_folder in os.listdir(input_dir):
    if genre_folder.startswith('.'):
        continue
    
    genre_dir = os.path.join(input_dir, genre_folder)
    output_genre_dir = os.path.join(output_dir, genre_folder)
    
    # create the output folder if it does not exist
    if not os.path.exists(output_genre_dir):
        os.makedirs(output_genre_dir)
    
    for song_folder in os.listdir(genre_dir):
        if song_folder.startswith('.'):
            continue
        
        song_dir = os.path.join(genre_dir, song_folder)
        
        # Load bass and drum audio and sum them
        bass_audio = MonoLoader(filename=os.path.join(song_dir, 'bass.mp3'))()
        drum_audio = MonoLoader(filename=os.path.join(song_dir, 'drums.mp3'))()
        mixed_audio = bass_audio + drum_audio
        
        # Write mixed audio to output file
        output_file = os.path.join(output_genre_dir, f"{song_folder}.Percu.mp3")
        MonoWriter(filename=output_file)(mixed_audio)
```

###Building Harmonic Dataset

```python
#Building Percussive Dataset

# set input and output directories
input_dir = '/content/drive/MyDrive/GTZAN/Demux'
output_dir = '/content/drive/MyDrive/GTZAN/HARMO'

if not os.path.exists(output_dir):
  os.makedirs(output_dir)

# iterate through the files in the input directory
for genre_folder in os.listdir(input_dir):
    if genre_folder.startswith('.'):
        continue
    
    genre_dir = os.path.join(input_dir, genre_folder)
    output_genre_dir = os.path.join(output_dir, genre_folder)
    
    # creamos la carpeta de salida si no existe
    if not os.path.exists(output_genre_dir):
        os.makedirs(output_genre_dir)
    
    for song_folder in os.listdir(genre_dir):
        if song_folder.startswith('.'):
            continue
        
        song_dir = os.path.join(genre_dir, song_folder)
        
        # Load bass and drum audio and sum them
        other_audio = MonoLoader(filename=os.path.join(song_dir, 'other.mp3'))()
        vocals_audio = MonoLoader(filename=os.path.join(song_dir, 'vocals.mp3'))()
        mixed_audio = other_audio + vocals_audio
        
        # Write mixed audio to output file
        output_file = os.path.join(output_genre_dir, f"{song_folder}.Harmo.mp3")
        MonoWriter(filename=output_file)(mixed_audio)
```

```python
# Define the source and destination directories
source_dir = '/content/drive/MyDrive/GTZAN/genres_original'
dest_dir = '/content/drive/MyDrive/GTZAN/genres_original-NO-CLASSIC'

# Get the list of files and folders in the source directory
files = os.listdir(source_dir)

# Loop over the files and folders
for file in files:
    # Check if the current item is a folder and if it is not the 'classical' folder
    if os.path.isdir(os.path.join(source_dir, file)) and file != 'classical':
        # If it is a folder and not 'classical', copy the folder to the destination directory
        shutil.copytree(os.path.join(source_dir, file), os.path.join(dest_dir, file))
    elif os.path.isfile(os.path.join(source_dir, file)):
        # If it is a file, copy the file to the destination directory
        shutil.copy2(os.path.join(source_dir, file), dest_dir)
```
